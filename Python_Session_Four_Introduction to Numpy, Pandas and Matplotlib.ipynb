## **1Ô∏è‚É£ Introduction to NumPy**
### **üîπ What is NumPy?**
**NumPy** is a powerful Python library for numerical computing. It provides support for multi-dimensional arrays and matrices, and comes with a variety of mathematical functions to operate on these arrays.

**Key Features of NumPy:**
- Efficient storage and manipulation of large arrays.
- Element-wise operations (addition, subtraction, etc.) on arrays.
- Linear algebra operations.
- Random number generation and statistics functions.

```python
import numpy as np

# Example of creating a 1D NumPy array and performing element-wise operation
arr = np.array([1, 2, 3, 4, 5])
print(f"1D Array: {arr}")

# Element-wise operation (square of each element)
arr_squared = arr ** 2
print(f"Squared Array: {arr_squared}")
```

### **üîπ Example of 2D NumPy Array**
You can also create 2D arrays (matrices) and perform operations on them.


```python
# Creating a 2D NumPy array (Matrix)
matrix = np.array([[1, 2], [3, 4], [5, 6]])
print(f"2D Matrix: 
{matrix}")

# Matrix multiplication (dot product)
matrix_mult = np.dot(matrix, matrix.T)
print(f"Matrix Multiplied by its Transpose: 
{matrix_mult}")
```

## **2Ô∏è‚É£ Introduction to Pandas**
### **üîπ What is Pandas?**
**Pandas** is a data manipulation and analysis library, specifically designed for working with tabular data (like spreadsheets or SQL databases). It provides powerful, flexible **DataFrame** and **Series** objects for working with structured data.

**Key Features of Pandas:**
- Handling and manipulation of structured data (tabular).
- Handling missing data.
- Data cleaning and transformation.
- Grouping, sorting, and filtering.

### **üîπ Creating a Pandas DataFrame**
Now, let's create a DataFrame using data from NumPy arrays. Pandas allows us to easily structure data in tabular form.

```python
import pandas as pd

# Creating a DataFrame from NumPy arrays
data = {
    'MathScore': np.array([88, 92, 79, 85, 94]),
    'ReadingScore': np.array([89, 96, 72, 85, 97]),
    'WritingScore': np.array([91, 87, 78, 84, 93])
}

df = pd.DataFrame(data)
df
```

### **üîπ Example of Basic Operations on DataFrame**
You can perform various operations like filtering, sorting, and accessing data from DataFrames.

## **3Ô∏è‚É£ Introduction to Matplotlib**
### **üîπ What is Matplotlib?**
**Matplotlib** is a plotting library for Python, providing functions to create a wide variety of visualizations like line plots, bar charts, histograms, and more.

**Key Features of Matplotlib:**
- Creating various types of plots (line, bar, scatter, etc.).
- Customization (titles, labels, legends, etc.).
- Integration with other libraries like Pandas for visualization.

```python
import matplotlib.pyplot as plt

# Data for plotting
x = df['MathScore']
y = df['ReadingScore']

# Plotting a scatter plot to visualize Math vs. Reading Scores
plt.scatter(x, y)
plt.title('Math vs. Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

### **üîπ Creating a Bar Plot**
A bar plot is useful for comparing quantities across different categories.


```python
# Data for bar plot
x = ['A', 'B', 'C', 'D']
y = [10, 15, 7, 20]

# Plotting the bar graph
plt.bar(x, y)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

### **üîπ Creating a Pie Chart**
Pie charts are useful for showing proportions of a whole.


```python
# Data for pie chart
labels = ['Math', 'Reading', 'Writing']
sizes = [35, 35, 30]

# Plotting the pie chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Performance Distribution by Subject')
plt.show()
```

### **üîπ Creating a Histogram**
Histograms are used for visualizing the distribution of numerical data.


```python
# Data for histogram
scores = df['MathScore']

# Plotting the histogram
plt.hist(scores, bins=5, edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

## **4Ô∏è‚É£ Combining NumPy, Pandas, and Matplotlib**
In real-world applications, we often combine **NumPy** for numerical operations, **Pandas** for data manipulation, and **Matplotlib** for visualization. Below is an example that demonstrates all three libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Creating a DataFrame with Pandas
data = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Revenue': [100, 150, 200, 250, 300]
}

df = pd.DataFrame(data)

# Perform a simple NumPy operation (growth rate calculation)
df['Growth Rate'] = np.diff(df['Revenue'], prepend=df['Revenue'][0]) / df['Revenue'][0] * 100

# Plotting the data with Matplotlib
plt.plot(df['Year'], df['Revenue'], label='Revenue')
plt.plot(df['Year'], df['Growth Rate'], label='Growth Rate')

plt.title('Revenue and Growth Rate Over Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```

## Visualizations

```python
df = df_exams
```

```python
# Create new columns based on the dataset
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
df['average_score'] = df['total_score'] / 3
df['pass_fail'] = df['total_score'].apply(lambda x: 1 if x >= 180 else 0)

# Show the new dataset with the added columns
df.head()
```

```python
df.shape
```

```python
# Bar plot for the distribution of total scores
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['total_score'], color='skyblue')
plt.title('Distribution of Total Scores')
plt.xlabel('Student Index')
plt.ylabel('Total Score')
plt.show()
```

```python
# Scatter plot comparing Math and Reading scores
plt.figure(figsize=(10, 6))
plt.scatter(df['math score'], df['reading score'], color='orange')
plt.title('Math vs. Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```

```python
# Box plot showing the distribution of scores for each subject
plt.figure(figsize=(10, 6))
plt.boxplot([df['math score'], df['reading score'], df['writing score']], labels=['Math', 'Reading', 'Writing'])
plt.title('Score Distribution by Subject')
plt.ylabel('Score')
plt.show()
```

```python
# Histogram for Math Scores distribution
plt.figure(figsize=(10, 6))
plt.hist(df['math score'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Pie chart for Pass/Fail distribution
pass_fail_counts = df['pass_fail'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(pass_fail_counts, labels=['Fail', 'Pass'], autopct='%1.1f%%', startangle=90, colors=['red', 'green'])
plt.title('Pass/Fail Distribution')
plt.show()
```

```python
# Stacked bar chart for subject scores by Gender
gender_grouped = df.groupby('gender')[['math score', 'reading score', 'writing score']].sum()
gender_grouped.plot(kind='bar', stacked=True, figsize=(10, 6), color=['skyblue', 'orange', 'green'])
plt.title('Total Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Score')
plt.show()
```

```python
import seaborn as sns

# Compute the correlation matrix for Math, Reading, and Writing scores
corr_matrix = df[['math score', 'reading score', 'writing score']].corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix of Scores')
plt.show()
```

```python
# Line chart for average scores over the student index
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['average_score'], color='purple', marker='o', linestyle='dashed')
plt.title('Average Scores of Students')
plt.xlabel('Student Index')
plt.ylabel('Average Score')
plt.show()
```
